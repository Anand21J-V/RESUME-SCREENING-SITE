```python
# REMOVING ALL THE SPECIAL CHARACTERS FROM THE DATASET

import re
def cleanResume(txt):
    cleanText = re.sub('http\S+\s', ' ', txt)
    cleanText = re.sub('RT|cc', ' ', cleanText)
    cleanText = re.sub('#\S+\s', ' ', cleanText)
    cleanText = re.sub('@\S+', '  ', cleanText)  
    cleanText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText) 
    cleanText = re.sub('\s+', ' ', cleanText)
    return cleanText
```

```python
# TESTING THE ABOVE CODE

cleanResume("my #### $ #  #noorsaeed webiste like is this http://heloword and access it @gmain.com")
```

```python
# NOW, APPLYING THE ABOVE CODE TO THE DATASET AND MAKING A NEW FEATURE OF THE NAME RESUME

df['Resume'] = df['Resume'].apply(lambda x: cleanResume(x))
```

```python
df['Resume'][0] 
```

## NOW...CONVERTIG THE CATEGORICAL FEATURES INTO NUMERICAL FEATURES

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(df['Category'])
df['Category'] = le.transform(df['Category'])
```

```python
df.Category.unique()
```

## NOW...CONVERTING THE RESUME FEATURE INTO VECTORS

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')

tfidf.fit(df['Resume'])
requredTaxt  = tfidf.transform(df['Resume'])
```

## SPLITTING BOTH THE FEATURES

```python
# NOW SPLITTING THE FEATURE INTO TRAIN AND TEST MODEL

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(requredTaxt, df['Category'], test_size=0.2, random_state=42)
```

```python
# CHECKING THE SHAPE OF THE SPLITTED TRAIN AND TEST DATASET

print("Shape of the X train dataset :",X_train.shape)
print("Shape of the Y train dataset :",y_train.shape)
print("Shape of the X test dataset :",X_test.shape)
print("Shape of the Y tesst dataset :",y_test.shape)
```

## NOW...TRAINING THE MODEL AND PRINT THE CLASSIFICATION REPORT

```python
# IMPORTING THE LIBRARIES FOR TRAINING THE MODEL

from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score

# TRAINING THE MODEL OF THE DATASET

clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train,y_train)
ypred = clf.predict(X_test)
print((accuracy_score(y_test,ypred))
```

```python
# MODEL GIVING 98% OF ACCURACY

print("Accuracy of the model :",0.9844559585492227*100)
```
